```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
#1: unreliable
#0: reliable
data=pd.read_csv('./../../data/all_data_updated.csv')
data.info()
cleanup_nums = {"type":     {"fake": 1, "real": 0}}
data.replace(cleanup_nums, inplace=True)
np.random.seed(seed=42)
print(data)
```

```python
from sklearn.model_selection import train_test_split
#data.type
data=data.iloc[:5000,:]
print(data)
y = data['type']
#X = data.drop('type',axis =1)
from sklearn.feature_extraction.text import TfidfVectorizer
data['total']=data['title']+' '+data['author']+data['text']
v = TfidfVectorizer()
x = v.fit_transform(data['total'])
#print(data)
#data['totalVector']=list(x.toarray())
df1 = pd.DataFrame(x.toarray(), columns=v.get_feature_names())
# 3. concatenate the tfidf dataframe to the original one 
data_updated = pd.concat([data, df1], axis=1)
#print(data_updated['total'])
X=data_updated.drop('total',axis=1)
#X=data_updated.iloc[:,12:]
export_csv = data_updated.to_csv (r'C:\Users\Harsh\Desktop\export_dataframe.csv', index = None, header=True)
#data.head()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X.head(3)
```

```python
#print(data['totalVector'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold


regressor = LogisticRegression()
lm_2 = regressor.fit(X_train, y_train)
cv = KFold(n_splits=5, shuffle=True, random_state=42)

pred_train_y = regressor.predict(X_train)
pred_test_y = regressor.predict(X_test)
cv_score = np.mean(cross_val_score(lm_2, X, y, cv=cv, scoring='r2'))

print ('%8s %8s %8s %8s' % ('Train R^2','Test R^2', 'CV R^2', 'MSE'))
print('%.6f %.6f %.6f %.6f'%  (lm_2.score(X_train,y_train), lm_2.score(X_test,y_test), cv_score,
                               np.mean((pred_test_y -y_test)**2)))
```
